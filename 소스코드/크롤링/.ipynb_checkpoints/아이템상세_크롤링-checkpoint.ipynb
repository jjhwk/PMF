```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from urllib.request import urlopen
from selenium.webdriver.common.keys import Keys
import time
import urllib.request
import pandas as pd
# import pymysql
```

```python
df = pd.read_csv("./woman_all_info.csv")
```

```python
df.head()
```

```python
ndf = df[["2", "3", "4", "5", "6", "7"]]
ndf.tail()
```

```python
ndf.tail()
```

```python
clothes_info = []
def clothes(id):
        # 옷 상세페이지 접속하기
        url = "https://www.musinsa.com/app/goods/"

        driver = webdriver.Chrome()
        
        try:
            driver.get(url + str(id)) # url 접속

            soup = BeautifulSoup(driver.page_source, "lxml")

            # 옷 종류 가져오기 
            category = soup.select("div a.product-detail__sc-up77yl-1")
            
            time.sleep(0.1)

            # 옷 이름 가져오기
            clothes_name = soup.find("h3", class_ = "product-detail__sc-1klhlce-3")
            
            time.sleep(0.1)

            # 옷 이미지
            images = soup.find("img", class_ = "product-detail__sc-p62agb-2")
            
            time.sleep(0.1)

            driver.close()

            # 가져온 옷 종류 리스트로 담기    
            clothes_list = []
            main_category = []
            sub_category = []

            # 옷 종류 추출하기
            for i in category:
                clothes_list.append(i.string)

            # 대분류 소분류 나누기
            for i in clothes_list:
                main_category.append(clothes_list[0]) 
                sub_category.append(clothes_list[1])

            print(clothes_list[0])
            print(clothes_list[1])

            # 옷 제목 추출하기
            clothes_title = []

            clothes_title.append(clothes_name.string)

            print(clothes_title[0])

            # 이미지 주소 

            image_urls = []
            image_urls.append(images.get("src"))
            print(image_urls[0])
            clothes_info.append([id, clothes_list[0], clothes_list[1],clothes_title[0], image_urls[0]])
        
        except:
            pass
```

```python
# 데이터프레임과 내부에 들어있는 값 집어넣고 돌리면
# 해당 컬럼과 위치 알려줌

def find_value(df, value):
    result = df.applymap(lambda x: x == value)  # 각 요소가 찾는 값과 일치하는지 확인
    rows, cols = result.any(axis=1), result.any(axis=0)  # 행과 열 각각에 대해 True 값이 있는지 확인

    found_indices = []
    for row in df.index[rows]:  # True 값을 가진 행을 반복
        for col in df.columns[cols]:  # True 값을 가진 열을 반복
            if df.at[row, col] == value:  # 값이 일치하는지 다시 확인
                found_indices.append((row, col))

    return found_indices

find_value(ndf,1375408.0)
```

# 데이터 수집

```python
# 오래 걸리니 컴퓨터 나눠서 돌리기 
# ndf 컬럼명, 2번째 for문 range값만 수정해서 돌리기
a = []

for i in ndf[["2"]]: # 2번 컬럼만 수집
    q = ndf[i]
    
    for j in range(15035, 15934): # 끊긴 인덱스부터 끝까지
        o = q[j]
        
        try:
            w = int(o)

            if w not in a:
                a.append(w)

        except:
            pass

        
for id in a:
    clothes(id)
```

```python
# 수집된 데이터 프레임으로 변환
# 중간에 끊을때는 꼭 저장하기

all_clothes_info = pd.DataFrame(clothes_info)
all_clothes_info.head()
```

```python
all_clothes_info.to_csv("./data/women_model_item_2(15035_).csv", index = False)
```
